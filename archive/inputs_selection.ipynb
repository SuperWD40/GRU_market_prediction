```python
import pandas as pd
import os
```

```python
df = pd.read_csv(os.getcwd()+'\\train_result\\GRU-BTCUSDT-1m-inputs_selection.csv')
df = df.drop(['Dataset size', 'Window size', 'Hidden size', 'Num layers', 'Input size', 'Dropout', 'Train size', 'Val size', 'Test size', 'Batch size', 'Epochs', 'Learning rate', 'Training device', 'Outputs', 'Unnamed: 0'], axis=1)
```

```python
numeric_cols = df.select_dtypes(include=['number'])
```

```python
input_columns = [f"Input_{i}" for i in range(14)]
df_melted = df.melt(id_vars=numeric_cols.columns, value_vars=input_columns, var_name="Input_Type", value_name="Input_Value")
```

```python
df_melted.groupby("Input_Value")[numeric_cols.columns].quantile(0.75)
```

```python
df_melted.to_csv(os.getcwd()+'\\train_result\\GRU-BTCUSDT-1m-inputs_selection-melted.csv')
```
